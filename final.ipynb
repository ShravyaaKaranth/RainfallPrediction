```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
```

```python
df1=pd.read_csv("/content/india_monthly_rainfall_data.csv",engine="python")
df1.head()
```

```python
#df_new = df1[df1['State'] == 'Karnataka']
# da = df1['State'] == 'Dadra & Nagar Haveli'
da = df1['State'] == 'Andhra Pradesh'
df_new = pd.DataFrame(df1[da])
print(df_new)
# daa = df_new['District'] == 'Hyderabad'
# df_newnew = pd.DataFrame(df_new[daa])
# print(df_newnew)
```

```python
# df_newnew.info()
df_new.info()
```

```python
annual = []
jtod = df_new.loc[:, ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec')]
# print(jtod.head())

# print((jtod.iloc[0]).sum())

for i in range((len(df_new.index))):
  a = ((jtod.iloc[i]).sum())/12
  # b = a/12
  annual.append(a)
```

```python
# print(annual[0:10])
```

```python
df_new['Annual'] = annual
df_new.head()
```

```python
#p_list = sn["Precipitation"].tolist()
```

```python
def norm(df_name, column_name):
  p_list = df_name[column_name].tolist()
  norm_p= []
  for i in df_name[column_name].values:
    z = (i - min(p_list))/ (max(p_list) - min(p_list))
    norm_p.append(z)
  # print(min(p_list))
  # print(max(p_list))
  return norm_p

# p_list = sn["Precipitation"].tolist()
# norm_p= []
# for i in sn['Precipitation'].values:
#   z = (i - min(p_list))/ (max(p_list) - min(p_list))
#   norm_p.append(z)
```

```python
Jan_norm = norm(df_new, 'Jan')
Feb_norm = norm(df_new, 'Feb')
Mar_norm = norm(df_new, 'Mar')
Apr_norm = norm(df_new, 'Apr')
May_norm = norm(df_new, 'May')
Jun_norm = norm(df_new, 'Jun')
Jul_norm = norm(df_new, 'Jul')
Aug_norm = norm(df_new, 'Aug')
Sep_norm = norm(df_new, 'Sep')
Oct_norm = norm(df_new, 'Oct')
Nov_norm = norm(df_new, 'Nov')
Dec_norm = norm(df_new, 'Dec')
Ann_norm = norm(df_new, 'Annual')
```

```python
anan = pd.DataFrame()

anan['Jan'] = Jan_norm
anan['Feb'] = Feb_norm
anan['Mar'] = Mar_norm
anan['Apr'] = Apr_norm
anan['May'] = May_norm
anan['Jun'] = Jun_norm
anan['Jul'] = Jul_norm
anan['Aug'] = Aug_norm
anan['Sep'] = Sep_norm
anan['Oct'] = Oct_norm
anan['Nov'] = Nov_norm
anan['Dec'] = Dec_norm
anan['Annual'] = Ann_norm



anan.head()
```

```python
from sklearn.model_selection import train_test_split
y= anan['Annual']
x = anan.loc[:, ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec')]
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2, 
                                                    random_state=0)
# print(x_train.head())
```

```python
model = MLPRegressor()
model.fit(x_train, y_train)
# print(model)
```

```python
expected_y  = y_test
predicted_y = model.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.r2_score(expected_y, predicted_y))#r2 value
print(metrics.mean_squared_log_error(expected_y, predicted_y))#rmse
```
